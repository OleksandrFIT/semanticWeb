```python
from SPARQLWrapper import SPARQLWrapper, JSON

# Ініціалізація SPARQL endpoint
sparql = SPARQLWrapper("http://dbpedia.org/sparql")

# SPARQL запит
sparql.setQuery("""
    PREFIX dbr: <http://dbpedia.org/resource/>
    PREFIX dbo: <http://dbpedia.org/ontology/>
    PREFIX dbp: <http://dbpedia.org/property/>

    SELECT ?diseaseLabel ?cause
    WHERE {
        ?diseaseLabel a dbo:Disease ;
           dbp:field dbr:Gastroenterology ;
           dbo:medicalCause ?cause .
    }
""")

# Вказуємо, що очікуємо результат у форматі JSON
sparql.setReturnFormat(JSON)

# Виконання SPARQL запиту
results = sparql.query().convert()

# Виведення назви захворювань та їх причин
for result in results["results"]["bindings"]:
    disease_label = result["diseaseLabel"]["value"]
    cause = result["cause"]["value"]
    print("Захворювання:", disease_label)
    print("Причина:", cause)
    print()
```
